```python
import numpy as np
from matplotlib import pylab as plt
```

```python
np.random.seed(302)
X = (np.random.randn(100) * 3.5) + 12
```

```python
X.mean()
```

```python
plt.hist(X, bins = 10)
plt.show()
```

```python
X_minmax = (X - X.min())
plt.hist(X_minmax, bins = 10)
plt.show()
```

```python
X_minmax = X_minmax / X_minmax.max()
plt.hist(X_minmax, bins = 10)
plt.show()
```

```python
X_standardized = (X - X.mean()) / X.std()
plt.hist(X_standardized, bins = 10)
plt.show()
```
